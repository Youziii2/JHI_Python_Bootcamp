## Pandas

```python
import pandas as pd
```

#### 1. Create a pandas DataFrame named `df` using the following table of names, ages, and scores.

| Name  | Age | Score |
|-------|-----|-------|
| Anna  | 23  | 88    |
| Ben   | 35  | 92    |
| Cara  | 29  | 79    |
| David | 41  | 85    |

```python
df = pd.DataFrame({'Name': ['Anna', 'Ben', 'Cara', 'David'],
    'Age': [23, 35, 29, 41],
    'Score': [88, 92, 79, 85]})
print(df)
```

#### 2. Use pandas methods on `df` to find the dataset’s dimensions (number of rows and columns) and calculate the average score.

```python
print(df.shape)
average = df['Score'].mean()
print(average)
```

#### 3. Filter the DataFrame `df` to keep only rows where the score is greater than or equal to 85 **and** the age is greater than 30.


```python
filtered = df[(df['Score'] >= 85) & (df['Age'] > 30)]
print(filtered)
```

#### 4. Create a new column `Score_decimal` by dividing the `Score` column by 100.

```python
df['Score_decimal'] = df['Score'] / 100
print(df)
```

## Matplotlib

```python
import matplotlib.pyplot as plt
```

#### 1. **Bar Plot.** Create a bar chart showing each person’s Name on the x-axis and their Score on the y-axis.

```python
plt.bar(df['Name'], df['Score'])
plt.xlabel('Name')
plt.ylabel('Score')
plt.title('Score by Name')
plt.show()
```

#### 2. **Scatter Plot.** Create a scatter plot for df showing the relationship between `Age ` and `Score`.

```python
plt.scatter(df['Age'], df['Score'])
plt.xlabel('Age')
plt.ylabel('Score')
plt.title('Age V.S. Score')
plt.show()
```

#### 3. **Histogram**. Create a histogram to show the distribution of the `Score` column.

```python
plt.hist(df['Score'])
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Distribution of Scores')
plt.show()
```
