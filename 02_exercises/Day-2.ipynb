## Strings

#### 1. Given string contains a combination of the lower and upper case letters. Write a program to arrange the characters of a string so that all lowercase letters should come first.
- Ex. PyNaTive --> yaivePNT

*Hint! Iterate each character from a string and check if the current character is the lower or upper case using `islower()` string function*

```python
string = "PyNaTive"
lower = ''
upper = ''
for c in string:
    if c.islower():
        lower = lower + c
    else:
        upper = upper + c
result = lower + upper
print(result)
```

#### 2. Write a program to check if two strings are balanced. For example, strings s1 and s2 are balanced if all the characters in the s1 are present in s2. The character’s position doesn’t matter.

- Ex. s1 = "Yn" , s2 = "PYnative" True, since Yn is in s2
- Ex. s1 = "Ynf", s2 = "PYnative" False, since f is not in s2

*Hint! Iterate each character from a string s1 and check if the current character is present in the string s2.*

```python
s1 = "Yn"
s2 = "PYnative"

balanced = all(c in s2 for c in s1)
print(balanced)
```
